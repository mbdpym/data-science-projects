1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

import warnings
warnings.filterwarnings("ignore")

renda = pd.read_csv('previsao_de_renda.csv')
renda.head()
````

````python
ax = sns.barplot(x="posse_de_veiculo", y="renda", hue = 'posse_de_veiculo',
                   data=renda, dodge=False, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'posse_de_veiculo',
                 data=renda)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Podemos ver com os dois gráficos que pessoas com veículo tendem a ter uma renda maior

````python
ax = sns.barplot(x="posse_de_imovel", y="renda", hue = 'posse_de_imovel',
                 data=renda, dodge=False, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'posse_de_imovel',
                 data=renda, dodge=False, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Pessoas com posse de imóvel apresentam renda maior, mas não por muito
